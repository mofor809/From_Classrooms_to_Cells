```python
#able to see the likelihood for charge type by race
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Define custom colormap
custom_colors = ["#ADD8E6", "yellow", "#FF3278"]
cmap = LinearSegmentedColormap.from_list("custom_cmap", custom_colors)

# Crosstab and heatmap
ct = pd.crosstab(df["RACE"], df["CHARGE TYPE"], normalize='index')
sns.heatmap(ct, annot=True, cmap=cmap)

# Labels and title
plt.rcParams['font.family'] = 'serif'
plt.title("Likelihood of Charge Type by Race",fontname='serif', fontsize=20)
plt.ylabel("Race")
plt.xlabel("Charge Type")

plt.savefig("charge_type_by_race.png", dpi=300, bbox_inches='tight')

plt.show()


#felonies are known to be harsher than misdeanors becasue they carry mroe severe and long term punishments
#sentences for felonies are usually more than a year while a misdeamonor is usually up to one year in county jail
#interesting to see how the likelihood for Black chidlren in chicago to get a misdeamor is the higest out of every race
#while the likehood for white children adn children of other races get a misdemeanor is higher than other races
```
